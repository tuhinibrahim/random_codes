````
import re

def extract_dates(file_path):
    # Regular expression to match different date formats
    date_patterns = [
        r'\b(?:January|February|March|April|May|June|July|August|September|October|November|December) \d{4}\b']

    extracted_dates = []

    with open("/content/16,500 murders in the last five years of AL.txt", 'r', encoding='utf-8') as file:
        content = file.read()

        for pattern in date_patterns:
            matches = re.findall(pattern, content)
            extracted_dates.extend(matches)

    return extracted_dates

# Example usage
file_path = "yourfile.txt"  # Replace with your actual file path
dates_found = extract_dates(file_path)

if dates_found:
    print("Extracted Dates and Months:")
    for date in dates_found:
        print(date)
else:
    print("No dates found in the file.")
````

````
import re

def extract_murdered_count(file_path):
    # Regular expression to match numbers followed by "murdered", "killed", "dead", etc.
    pattern = r'\b(\d{1,3}(?:,\d{3})*)\s*(?:murdered|killed|dead)\b'

    extracted_numbers = []

    with open('/content/16,500 murders in the last five years of AL.txt', 'r', encoding='utf-8') as file:
        content = file.read()

        # Find all matches of the pattern
        matches = re.findall(pattern, content, re.IGNORECASE)

        # Convert matched strings to integers
        for match in matches:
            # match[0] is the number as a string, remove commas and convert to int
            number = int(match[0].replace(",", ""))
            extracted_numbers.append(number)

    return extracted_numbers

# Example usage
file_path = "yourfile.txt"  # Replace with your actual file path
numbers_found = extract_murdered_count(file_path)

if numbers_found:
    print("Extracted Numbers of People Murdered/Killed:")
    for number in numbers_found:
        print(number)
else:
    print("No numbers related to murdered/killed people found in the file.")
````

````
import re

def extract_numbers_from_paragraph(file_path, paragraph_title):
    # Regular expression to match all numbers (including those with commas)
    number_pattern = r'\b\d{1,3}(?:,\d{3})*\b'

    numbers = []

    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()

        # Split the content into paragraphs based on a specific pattern (e.g., paragraph titles or markers)
        paragraphs = content.split('\n\n')  # Assuming paragraphs are separated by blank lines

        # Loop through the paragraphs to find the one containing the paragraph title
        for paragraph in paragraphs:
            if paragraph_title in paragraph:
                # Extract all numbers from the identified paragraph
                numbers = re.findall(number_pattern, paragraph)
                break  # Once the paragraph is found, stop searching

    return numbers

# Example usage
file_path = "/content/16,500 murders in the last five years of AL (1).txt"  # Replace with your actual file path
paragraph_title = "During the July Uprising"  # Title of the paragraph to extract numbers from
numbers_found = extract_numbers_from_paragraph(file_path, paragraph_title)

if numbers_found:
    print(f"Extracted Numbers from '{paragraph_title}' Paragraph:")
    for number in numbers_found:
        print(number)
else:
    print(f"No numbers found in the paragraph '{paragraph_title}'.")
````

````
import pandas as pd

def load_rent_data(file_path):
    """Loads the CSV file into a DataFrame."""
    try:
        df = pd.read_csv(file_path)
        print(df)
        return df
    except FileNotFoundError:
        print("Error: File not found. Please check the file path.")
        return None

def get_rent_difference(df, branch1, branch2):
    """Finds the rent difference between two branches."""
    # Convert column names to lowercase for consistency
    #df.columns = df.columns.str.lower()

    # Ensure necessary columns exist
    required_columns = {"BRANCH", "MONTHLY RENT"}
    if not required_columns.issubset(set(df.columns)):
        print("Error: CSV file must contain 'BRANCH' and 'MONTHLY RENT' columns.")
        return None

    # Find the rent values for both branches
    rent1 = df[df["NOP"].str.lower() == branch1.lower()]["MONTHLY RENT"].values
    rent2 = df[df["BRANCH"].str.lower() == branch2.lower()]["MONTHLY RENT"].values

    # Ensure both branches exist in the dataset
    if rent1.size == 0 or rent2.size == 0:
        print(f"Error: One or both branches not found in the dataset.")
        return None

    # Calculate the rent difference
    difference = abs(rent1[0] - rent2[0])

    print(f"MONTHLY RENT of '{branch1}': {rent1[0]}")
    print(f"MONTHLY RENT of '{branch2}': {rent2[0]}")
    print(f"Difference in Rent: {difference}")

# Example Usage
file_path = "/content/Advance Rent - Draft.csv"  # Replace with your actual file path
df = load_rent_data(file_path)

if df is not None:
    branch1 = input("Enter the first BRANCH: ")
    branch2 = input("Enter the second BRANCH: ")
    get_rent_difference(df, branch1, branch2)
````

````
import pandas as pd
from datetime import datetime
from dateutil import rrule

# Load CSV data
df = pd.read_csv('/content/Advance Rent - Draft.csv', thousands=',')  # Handles comma-formatted numbers

def calculate_total_payment():
    # Get branch input
    branch = input("Enter branch name: ").strip()

    try:
        monthly_rent = df[df['BRANCH'] == branch]['MONTHLY RENT'].values[0]
    except IndexError:
        print("Error: Branch not found")
        return

    # Calculate time period
    start_date = datetime.now().replace(day=1)  # First day of current month
    end_date = datetime(2030, 12, 31)           # End target date

    # Calculate number of months using monthly recurrence rule
    months = rrule.rrule(
        rrule.MONTHLY,
        dtstart=start_date,
        until=end_date
    ).count()

    # Calculate total payment
    total_payment = monthly_rent * months

    # Display results
    print(f"\nPayment Forecast for {branch}")
    print(f"Monthly Rent: {monthly_rent:,.2f}")
    print(f"Number of Months: {months} (from {start_date.strftime('%B %Y')} to December 2030)")
    print(f"Total Payment: {total_payment:,.2f}")

# Run the calculation
calculate_total_payment()
````

````
import random

# Predefined responses
responses = {
    "hello": ["Hello!", "Hi there!", "Hey! How can I help?"],
    "how are you": ["I'm just a bot, but I'm doing great!", "I'm good! How about you?"],
    "bye": ["Goodbye!", "See you soon!", "Take care!"],
    "what is your name": ["I'm ChatBot!", "You can call me Bot!", "I am your AI assistant."],
    "default": ["I'm not sure I understand.", "Can you rephrase that?", "Interesting! Tell me more."]
}

# Function to get response
def chatbot_response(user_input):
    user_input = user_input.lower()  # Convert input to lowercase
    for key in responses.keys():
        if key in user_input:
            return random.choice(responses[key])  # Return a random response for the matched key
    return random.choice(responses["default"])  # Default response if no match is found

# Chat loop
print("ChatBot: Hello! Type 'bye' to exit.")
while True:
    user_input = input("You: ")
    if user_input.lower() == "bye":
        print("ChatBot:", random.choice(responses["bye"]))
        break
    print("ChatBot:", chatbot_response(user_input))
````

````
class Car():
  """A simple attempt to represent a car."""
  def __init__(self, make, model, year):
    """Initialize attributes to describe a car."""
    self.make = make
    self.model = model
    self.year = year

    # Full capacity and level in gallons.
    self.fuel_capacity = 15
    self.fuel_level = 0

  def fill_tank(self):
    """Fill gas tank to capacity."""
    self.fuel_level = self.fuel_capacity
    print("Fuel tank is full.")

  def drive(self):
    """Simulate driving."""
    print("The car is moving.")

  def update_fuel_level(self, new_level):
    """Update the fuel level."""
    if new_level <= self.fuel_capacity:
      self.fuel_level = new_level
    else:
      print("The tank can't hold that much!")

  def add_fuel(self, amount):
    """Add fuel to the tank."""
    if (self.fuel_level + amount <= self.fuel_capacity):
      self.fuel_level += amount
      print("Added fuel.")
    else:
      print("The tank won't hold that much.")

my_car = Car('audi', 'a4', 2016)
print(my_car.make)
print(my_car.model)
print(my_car.year)

my_car.fill_tank()
my_car.drive()

my_car = Car('audi', 'a4', 2016)
my_old_car = Car('subaru', 'outback', 2013)
my_truck = Car('toyota', 'tacoma', 2010)
````

````
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance
# Private variable (can't be accessed directly)

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
# Controlled access through a method
        return self.__balance

# Create an object
account = BankAccount(1000)
account.deposit(500)

print(account.get_balance())
````

````
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "Some sound"

# Child class inheriting from Animal
class Dog(Animal):
    def speak(self):
        return "Bark"

class Cat(Animal):
    def speak(self):
        return "Meow"

dog = Dog("Buddy")
cat = Cat("Whiskers")

print(dog.name, ":", dog.speak())  # Output: Buddy : Bark
print(cat.name, ":", cat.speak())  # Output: Whiskers : Meow
````

````
class Bird:
    def fly(self):
        return "Some birds can fly"

class Sparrow(Bird):
    def fly(self):
        return "Sparrow flies high"

class Penguin(Bird):
    def fly(self):
        return "Penguins cannot fly"

# Using the same method but getting different results
birds = [Sparrow(), Penguin()]

for bird in birds:
    print(bird.fly())
# Different output based on the class
````

````
from abc import ABC, abstractmethod

class Vehicle(ABC):
  # Abstract class
    @abstractmethod
    def start_engine(self):
        pass
        # No implementation

class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"

class Bike(Vehicle):
    def start_engine(self):
        return "Bike engine started"

car = Car()
bike = Bike()

print(car.start_engine())
# Output: Car engine started
print(bike.start_engine())
# Output: Bike engine started
````

````
# Define the classes directly in your Colab notebook
class Car:
    def __init__(self, make, model, year):
        """Initialize attributes to describe a car."""
        self.make = make
        self.model = model
        self.year = year
        self.fuel_level = 100  # Assume full tank initially

    def fill_tank(self):
        """Fill the gas tank."""
        self.fuel_level = 100
        print("The gas tank is now full.")

    def drive(self):
        """Simulate driving."""
        print(f"The {self.year} {self.make} {self.model} is now driving.")

class ElectricCar(Car):
    def __init__(self, make, model, year):
        """Initialize attributes of the parent class and add electric-specific attributes."""
        super().__init__(make, model, year)  # Calls the parent constructor
        self.charge_level = 100  # Assume fully charged

    def charge(self):
        """Fully charge the vehicle."""
        self.charge_level = 100
        print("The vehicle is fully charged.")

    def drive(self):
        """Override drive method for electric cars."""
        print(f"The {self.year} {self.make} {self.model} is silently driving.")
````

````
import unittest

# Sample functions to test
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        raise ValueError("Cannot divide by zero")
    return a / b

# Test cases
class TestMathFunctions(unittest.TestCase):
    def test_add(self):
        self.assertEqual(add(2, 3), 5)

    def test_subtract(self):
        self.assertEqual(subtract(10, 5), 5)

    def test_multiply(self):
        self.assertEqual(multiply(4, 3), 12)

    def test_divide(self):
        self.assertEqual(divide(10, 2), 5)
        with self.assertRaises(ValueError):
            divide(10, 0)

if __name__ == "__main__":
    unittest.main()
````

````
def get_full_name(first, last, middle=''):
    """Return a full name, properly formatted."""
    if middle:
        full_name = "{0} {1} {2}".format(first, middle, last)
    else:
        full_name = "{0} {1}".format(first, last)
    return full_name.title()

class NamesTestCase(unittest.TestCase):
    """Tests for the 'full_names.py' module."""

    def test_first_last_name(self):
        """Do names like 'Janis Joplin' work?"""
        janis = get_full_name('janis', 'joplin', 'Cr7')
        self.assertEqual(janis, 'Janis Cr7 Joplin')

unittest.main(argv=[''], verbosity=2, exit=False)
````
